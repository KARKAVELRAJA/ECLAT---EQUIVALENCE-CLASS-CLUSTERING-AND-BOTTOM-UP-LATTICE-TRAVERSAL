## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Market_Basket_Optimisation.csv",header=None)
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***TRAINING THE DATASET***

```python
!pip install pyECLAT
```

```python
from pyECLAT import ECLAT

ECLAT_Model = ECLAT(data = data,verbose=True)
```

## ***DISPLAYING THE RESULT***

```python
result = ECLAT_Model.support()
result
```

## ***DISPLAYING THE RESULT IN DATAFRAME FORMAT***

```python
result_DataFrame = pd.DataFrame((pd.DataFrame([result]).values.reshape(120,1)),index=result,columns=["Support"])
result_DataFrame.index.name="Items"
result_DataFrame
```

## ***SORTING THE VALUES BASED ON THE SUPPORT COLUMN***

```python
result_DataFrame.nlargest(n=10, columns = "Support")
```
